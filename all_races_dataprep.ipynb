```python
#check if there is duplicates

def check_duplicates(df):
    if len(df) == len(set(df)):
        return False
    else:
        return True


result = check_duplicates(df)

if result:
    print('Yes, there is duplicates.')
else:
    print('No duplicates found.')

df.duplicated().sum()
```
